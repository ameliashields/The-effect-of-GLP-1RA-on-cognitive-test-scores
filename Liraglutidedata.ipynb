```python
import pandas as pd
import matplotlib.pyplot as plt
from google.colab import files
uploaded = files.upload()
df = pd.read_excel("vadini et al.xlsx")
print(df.head())
```

```python
def p_to_stars(p):
  if p < 0.001: return "***"
  elif p < 0.01: return "**"
  elif p < 0.05: return "*"
df["stars"] = df["P-value"].apply(p_to_stars)

plt.figure(figsize=(10,6))
plt.bar(df["Test Name"], df["% change"], yerr=df["SD"], capsize=6, color="blue")
plt.axhline(0, color="black", linewidth=0.8)
plt.xticks(rotation=35, ha="right")
plt.ylabel("% Change in Cognitive Test Scores")
plt.title("% Change in Cognitive Test Scores following Liraglutide Treatment (Vadini et al., 2020\)")

for i, (val, err, star) in enumerate(zip(df["% change"], df["SD"], df["stars"])):
  plt.text(i, val + err + 1, star, ha="center", va="bottom")

plt.tight_layout()
plt.show()
```
